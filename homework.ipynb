**Task:**

```

y: 7
z: 14

result:
```

````python
def retrieve_even_values(d):
    '''
    returns a new dictionary containing only the key-value pairs where the value is an even integer
    '''
    #a variable that is an empty dictionary
    new_dictionary = {}
    # a for loop to go through the original dictionary and look at the key and its value
    for k, v in d.items():
        # if the remainder of the value is 0 (if it's divisible by 2)
        if v % 2 == 0:
            # look at the key, take the value and put it into the new dictionary
            new_dictionary[k] = v
    # return the new dictionary once the loop is completed
    return new_dictionary
        
# Helper function
def print_dict(d):
    '''
    Iterate over a dictionary and print out each key and value.
    '''
    for k in d.keys():
        print(f'{k}: {d[k]}')

  


````

Write a function read_file_as_words(filename) that:

1. Opens a text file for reading

2. Reads the entire file as a single string

4. Returns a list of words in lowercase

**Sample Output:**

If the file contains:

```
Apple banana

```
Output:
```

```


````python
def read_file_as_words(filename):
  '''
  opens a text file, reads the file as a single string, splits the text into words, and returns a list of words lowercase
  '''
  # variable that opens the file for reading
  file = open(filename, 'r')
  # look at each line in the file
  for line in file:
    # split the line, giving multiple words
    word = line.split()
    # print the word in a list
    return word
  file.close()
  
````

Notice one blank line


````python
def get_data_row_count(filename):
  '''
  reads TSV file with a header row, returning the number of non-blank data rows
  '''
  # open file to read, then close the file
  with open(filename, 'r') as file:
    # reads the entire file and returns a list
    lines = file.readlines()
    # accumulator variable to count the lines
    count_lines = 0

    # for loop to go through each of the lines in the file, skipping the first element (header)
    for line in lines[1:]:
      # if it is not a blank line (there shouldn't be since this removed the blank lines)
      if line.strip():
        # then add the number of lines to the accumulator variable
        count_lines = count_lines + 1
    # return the total count of lines
    return count_lines

 

  
````

**Problem 8**

**Concept:** *Extracting one column from a CSV file.*

**Task:**

Write a function `get_column_data(filename, column_name)` that reads a CSV file with a header row and returns a list containing the values from the specified column.

If the column name is missing, the function should raise a `KeyError`.

**Sample Output:**

Input:
```
get_column_data("sales.csv", "price")
```
If column contains:
```
10.99
5.50
20.00
```
Output:
```
['10.99', '5.50', '20.00']
```
If column missing:
```
KeyError: Column 'discount' does not exist in file 'sales.csv'.
```


````python
import csv
def get_column_data(filename, column_name):
  ''' 
  reads a CSV file with a header row and returns a list containing the values from the specified column
  If the column name is missing, the function should rase a KeyError
  '''
  # an open list to put the values of the column that is asked for
  columns = []

  # open file to read, then close the file
  with open(filename) as file:
    # variable to read trhe entire file
    file_reader = csv.reader(file)
    # reads the header row
    header = next(file_reader)
    # if the column_name isn't in the header
    if column_name not in header:
      # print(f"KeyError: Column '{column_name}' does not exist in file {filename}.")
      # raise the KeyError and print the following error message
      raise KeyError(f"Column '{column_name}' does not exist in file '{filename}'")
    
    # taking the index of each column name in the header
    index_column = header.index(column_name)
    # for each row in the file
    for row in file_reader:
      # grab the values that correspond to the column_name and add then to the empty list
      columns.append(row[index_column])
  # return the completed columns list
  return columns
````

**Sample Output:**

```

```
Output:
```
[
 {'a': 1, 'b': 2, 'c': 3},
 {'a': 4, 'b': None, 'c': 6}

```


````python
def read_csv_to_dictionaries(filename):
  ''' 
  reads a CSV file with a header row and yields one dictionary per row
  '''
  # open file to read then closes it
  with open(filename, 'r') as file:
  # reads first line of line, strip ther spaces, split on comma
    headers = file.readline().strip().split(',')
    
    #loop through each row in the file
    for line in file:
      #extract list of fields
      values = line.strip().split(',')
      # build a dictionary, starting with an empty one
      d = {}

      # for each header for the length of the # of headers
      for i in range (len(headers)):
        # take the header and value of the index, and put it into the dictionary
        try:
          d[headers[i]] = int(values[i])
        # if there is no value with the header, make the value None
        except ValueError:
          d[headers[i]] = None
      # reads the data, returns the dictionary once loop is completed
      yield d

  # test code
  # print(values)
  # print(headers)
  # d = {headers[i] : values[i] for i in range (len(headers))}
  # print(d)
````

**Problem 10**

**Concept:** *Data cleaning + flagging “bad” rows (finagled).*

**Task:**

Write a generator function `parse_data_file(filename)` that reads a CSV file with a header row and yields dictionaries like Problem 9, except:

* Parse values as integers

* If parsing fails, store the value 3

* Add a key `"finagled"` to each dictionary only if:

    * `True` if at least one value failed to parse


**Sample Output:**

If CSV file:
```
a,b
5,7
8,abc
```
Output:
```
[
 {'a': 5, 'b': 7},
 {'a': 8, 'b': -1, 'finagled': True}
]
```


````python
def parse_data_file(filename):
  ''' 
  reads CSV file that parses values as intergers, if fails, store as 3, add key 'finagled' if True at least one valued didn't parse
  '''
  # open file to read then closes it
  with open(filename, 'r') as file:
    # reads first line of line, strip ther spaces, split on comma
    headers = file.readline().strip().split(',')
   
    #read through each line in the file
    for line in file:
      #extract list of fields
      values = line.strip().split(',')
      # build a dictionary, starting with an empty one
      d = {}
      # track the errors so we can change to finagled
      values_error = False

      # for each header for the length of the # of headers
      for i in range (len(headers)):
        # take the header and converted value of the index, and put it into the dictionary
        try:
          d[headers[i]] = int(values[i])
        # if the integer can't be parsed
        except ValueError:
          # set that header's value to -1
          d[headers[i]] = -1
          # change the values_error variable to True
          values_error = True
      # if we did catch and error
      if values_error:
        # add a new header name, making the value True
        d['finagled'] = True
      # return the completed dictionary
      yield d


````

**Task:**

Write a function `process_columns(input_filename, output_filename, column_names)` that:

1. Reads `input_filename` as a CSV with a header row

2. Writes `output_filename` as a CSV containing only the columns listed in `column_names`

3. The output columns must appear in the same order as `column_names`

Assume all requested `column_names` exist in the input.

**Sample Output:**

Input file:
```
name,age,grade
Alice,20,A
Bob,22,B
```
Function call:
```
process_columns("students.csv", "output.csv", ["name", "grade"])
```
Output file:
```
name,grade,
Alice,A,
Bob,B,
```


````python
def process_columns(input_filename, output_filename, column_names):
  ''' 
  reads input_filename as a csv with header row
  writes output_filename as a CSV containing only columns
  output columns appear in same order as column_names
  '''
  # open file, our input file
  with open(input_filename, 'r') as input_file:
    # variable to read the file
    input_reader = csv.reader(input_file)
    # variable for getting the header
    input_header = next(input_reader)
  
    # make an empty list for the columns indexes
    column_indexes = []
    # for each column name
    for column_name in column_names:
      # add the column names for each index and add it to the empty list, column_indexes
      column_indexes.append(input_header.index(column_name))

    # open file, our output file
    with open(output_filename, 'w') as output_file:
      # writing the output file
      output_writer = csv.writer(output_file)
      # write the header row
      output_writer.writerow(column_names)

      # write the data into the file
      for row in input_reader:
        # empty list to put in the data for each new row
        new_row = []
        # for each of the column indexes that were pulled into the column indexes new string
        for index in column_indexes:
          # append that column's data to the new row list
          new_row.append(row[index])
        # then write it into the new csv with the corresponding column
        output_writer.writerow(new_row)

      
  
   
````

````python
import json
def read_json_data(filename, key):
  ''' 
  prases file as JSON, if JSON object is a dictionary and contains key, return its value, otherwise return None
  '''
  #open file then close close
  with open(filename) as file:
    # variable to read the entire file
    text = file.read()
    # ​Reads the contents of a string as JSON
    json_data = json.loads(text)
  # print(json_data)

  try:
    # return the key 
    return json_data[key]
  # if there is no value of that name and will cause an error because it is missing, return None
  except (KeyError, TypeError):
    return None
````

**Problem 14**

**Concept:** *Reading a CSV file; computing derived values.*

**Task:**

Write a function `extract_trading_range(filename)` that reads a CSV file of stock prices.

For each row:

1. Extract `high` and `low`

2. Compute `trading_range = high - low`

3. Return a list of `trading_range` values (floats)

Assume the CSV has columns named exactly `high` and `low`.

**Sample Output:**

If CSV file:

```
date,high,low
2024-01-01,150,140
2024-01-02,200,180
```
Output:
```
[10.0, 20.0]
```


````python
def extract_trading_range(filename):
  ''' 
  reads CSV file of stock prices
  '''
  # empty list to put the tranding range values
  trading_ranges = []
  # opens file to read then closes it
  with open(filename, 'r') as file:
    #reads each row as a dictionary
    file_reader = csv.DictReader(file)
    
    # go through each row, skips header
    for row in file_reader:
      # take high values and turn them into floats
      high = float(row['high']) 
       # take low values and turn them into floats
      low = float(row['low'])
      # equation to get the trading range
      trading_range = high - low
      #add the range that is calculated to the list created
      trading_ranges.append(trading_range)
    # return with the new list
    return trading_ranges
````
